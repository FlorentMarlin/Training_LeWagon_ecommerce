```python
import pandas as pd
import numpy as np

df_order_reviews_dataset = pd.read_csv('/content/drive/MyDrive/Le Wagon - Data Analytics/30_Olist/olist_order_reviews_dataset.csv')
#https://drive.google.com/file/d/1VrvEI0nskovcyEmVGHDHvaAtQa8-4LSA/view?usp=drive_link
df_order_reviews_dataset
```

```python
s_filepath_output = '/content/drive/MyDrive/Le Wagon - Data Analytics/30_Olist/olist_order_reviews_dataset_en_2025-11-04.csv'
```

```python
from googletrans import Translator
```

```python
my_translator = Translator()
```

```python
ls_review_comment_en = []
i_count = 0
for s_review_comment_pt in df_order_reviews_dataset['review_comment_message']:
#for s_review_comment_pt in df_order_reviews_dataset['review_comment_message'][:10]:
  #print(s_review_comment_pt)
  #print(type(s_review_comment_pt))
  #if s_review_comment_pt is not None:
  #if not np.isnan(s_review_comment_pt):
  #if not s_review_comment_pt.isnull():
  if isinstance(s_review_comment_pt, str):
    #print(s_review_comment_pt)
    try:
      translation = my_translator.translate(s_review_comment_pt, src='pt', dest='en')
      translation = translation.text
    except Exception as e_translate:
      translation = 'ERROR'
      print('Failed to translate - ' + str(e_translate))
    ls_review_comment_en.append(translation)
    #print(translation.text)
  else:
    #print('None')
    ls_review_comment_en.append(None)
  i_count += 1
  if divmod(i_count, 100)[1] == 0:
    print(f'Done {i_count}')
  if divmod(i_count, 1000)[1] == 0:
    ls_review_comment_en_for_write = ls_review_comment_en + [None] * (df_order_reviews_dataset.shape[0] - len(ls_review_comment_en))
    df_order_reviews_dataset_en = df_order_reviews_dataset.copy(deep=True)
    df_order_reviews_dataset_en['review_comment_message_en'] = ls_review_comment_en_for_write
    df_order_reviews_dataset_en.to_csv(s_filepath_output, index=False)
    print(f'Done {i_count} - Saving')

#print(ls_review_comment_en)

ls_review_comment_en_for_write = ls_review_comment_en + [None] * (df_order_reviews_dataset.shape[0] - len(ls_review_comment_en))
df_order_reviews_dataset_en = df_order_reviews_dataset.copy(deep=True)
df_order_reviews_dataset_en['review_comment_message_en'] = ls_review_comment_en_for_write
#print(df_order_reviews_dataset_en)
df_order_reviews_dataset_en.to_csv(s_filepath_output, index=False)
```
